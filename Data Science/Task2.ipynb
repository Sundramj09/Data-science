```python
first = int(input('enter the 1st angle'))
second = int(input('enter the 2nd angle'))
third = int(input('enter the 3rd angle'))
if(first+second+third)==180 and first>0 and second>0 and third>0:
    print("forms a triangle")
else:
    print("does not form a triangle")
```

```python
cost_price = int(input('Enter cost price-'))
selling_price = int(input('Enter selling price-'))

if cost_price < selling_price:
    print('Profit')
elif cost_price > selling_price:
    print('Loss')
else:
    print('No Loss No Gain')
    
```

```python
menu= input("""
hi select an option
1. cm to ft
2. km to miles
3. usd to inr
4. exit
""")
if menu=='1':
    cm=float(input("enter the cm value"))
    print("ft value is",0.032*cm)
elif menu=='2':
    km=float(input("enter the km value"))
    print("miles value is:",0.62*km)
elif menu=='3':
    usd=float(input('enter usd'))
    print('inr',usd*80)
else:
    exit()
```

```python
n= int(input("enter the number"))
sum=0
i=1
while i<n+1:
    if i %5==0:
        i+=1
        continue
    sum+=i
    if sum>300:
        sum=sum-i
        break
        
    i+=1
print(sum)
        
```

```python
sum = 0
count = 0

while True:
  num = int(input('enter number'))
  if num == 0:
    break
  sum = sum + num
  count = count + 1

print('sum',sum)
print('avg',sum/count)
```

```python
L=[]
for i in range(2000,3200):
    if i % 7==0 and i % 5!=0:
        L.append(str(i))
    print(",".join(L))
```

```python
L = []
for i in range(1000,3001):
  flag = True

  curr = i

  while curr > 0:
    last = curr%10
    if last % 2 != 0:
      flag = False
      break
    curr = curr//10

  if flag == True:
    L.append(str(i))

print(",".join(L))
```
